```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from matplotlib.axes._axes import _log as matplotlib_axes_logger
from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_circles
```

```python
X,y=make_circles(100,factor=0.1,noise=0.1)
```

```python
X
```

```python
y
```

```python
plt.scatter(X[:,0],X[:,1], c=y ,s=50,cmap ='bwr')
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size = 0.20)
```

```python
X_train
```

```python
classifier1 = SVC(kernel='linear')
classifier1.fit(X_train,y_train)
```

```python
y_pred1 = classifier1.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred1)
```

```python
classifier2 = SVC(kernel='rbf')
classifier2.fit(X_train,y_train)
```

```python
y_pred2 = classifier2.predict(X_test)
```

```python
accuracy_score(y_test,y_pred2)
```

```python
classifier3 = SVC(kernel='poly')
classifier3.fit(X_train,y_train)
```

```python
y_pred3 = classifier3.predict(X_test)
```

```python
accuracy_score(y_test,y_pred3)
```

```python
classifier4 = SVC(kernel='sigmoid')
classifier4.fit(X_train,y_train)
```

```python
y_pred4 = classifier4.predict(X_test)
```

```python
accuracy_score(y_test,y_pred4)
```

```python
classifier = SVC(kernel='poly',degree=2)
classifier.fit(X_train,y_train)
y_pred = classifier.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
#decision boundary 

def plot_desicion_boundary(X,y,clf):
    X_set , y_set = X,y
    zero_one_colourmap = ListedColormap(['#FF0000', '#0000FF'])  # Red and Blue colors

    X1 , X2 = np.meshgrid(np.arange(start= X_set[:,0].min() - 1 ,
                                    stop = X_set[:,0].max() + 1 ,
                                    step = 0.01) ,
                          np.arange(start= X_set[:,1].min() - 1 ,
                                    stop = X_set[:,1].max() + 1 , 
                                    step = 0.01) )
    plt.contourf(X1,X2 ,clf.predict(np.array([X1.ravel(),
                                             X2.ravel()]).T).reshape(X1.shape) ,
                alpha = 0.75,
                cmap = zero_one_colourmap)

    plt.xlim(X1.min(),X1.max())            
    plt.ylim(X2.min(),X2.max())   

    for i ,j in enumerate (np.unique(y_set)):
        plt.scatter(X_set[y_set == j , 0] , X_set[y_set == j , 1] ,
                 c = (zero_one_colourmap)(i) , label = j)
    plt.title('SVM desicion boundary ')
    plt.xlabel('X1')
    plt.ylabel('X2')     
    plt.legend()
    return plt.show()
```

```python
plot_desicion_boundary(X,y,classifier1)
```

```python
plot_desicion_boundary(X,y,classifier2)
```

```python
plot_desicion_boundary(X,y,classifier3)
```

```python
plot_desicion_boundary(X,y,classifier4)
```

```python
def plot_3d_plot(X,y):
    r=np.exp(-(X**2).sum(1))
    ax= plt.subplot(projection='3d')
    ax.scatter3D(X[:,0],X[:,1],r,c=y , s=100,cmap='bwr')
    ax.set_xlabel('X1')
    ax.set_ylabel('X2')
    ax.set_zlabel('y')
    return ax
```

```python
plot_3d_plot(X,y)
```
